```python
import pandas as pd

df = pd.read_csv('labelled_embeddings.csv')

X = df.drop(columns=['label', 'class'])
X
```

```python
from sklearn.manifold import TSNE
t_sne = TSNE(n_components=2,
             learning_rate=200,
             init='pca',
             n_iter=10000,
             n_jobs=8,
             random_state=3,
             early_exaggeration=20)
X = t_sne.fit_transform(X)
X
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], s=1)
plt.show()
```

```python
y = df['class']
targets = list(set(y))
targets
```

```python
import matplotlib.colors as mcolors

colors = list(mcolors.XKCD_COLORS)[0:36]
colors
```

```python
colors = []
import numpy as np
for i in range(3):
    for j in range(3):
        for k in range(4):
            colors.append([(0.1 + i * 0.35, 0.1 + j * 0.35, 0.1 + k * 0.25)])
```

```python
finalDF = pd.DataFrame()
finalDF['T-SNE 1'] = X[:, 0]
finalDF['T-SNE 2'] = X[:, 1]
finalDF['target'] = df['class']

finalDF
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('T-SNE 1', fontsize = 15)
ax.set_ylabel('T-SNE 2', fontsize = 15)
ax.set_title('2 component T-SNE', fontsize = 20)
for target, color in zip(targets,colors):
    indicesToKeep = finalDF['target'] == target
    ax.scatter(finalDF.loc[indicesToKeep, 'T-SNE 1']
               , finalDF.loc[indicesToKeep, 'T-SNE 2']
               , c=color
               , s=20)

ax.legend(targets, bbox_to_anchor=(1, 0.25, 0.5, 0.5), loc='center left', ncol=2)
ax.grid()
fig.savefig('t-sne.png',bbox_inches='tight')
```
